```python
import os
from deepspeech import Model
from nltk.translate import Translator

# Initialize DeepSpeech model
deepspeech_model_path = "/content/drive/MyDrive/DARPG/DARPG_3/deepspeech-0.9.3-cp39-cp39-win_amd64.whl"
deepspeech_model = Model(deepspeech_model_path)

# Initialize NLTK translator
translator = Translator()

# Function to transcribe audio file to text
def transcribe_audio(audio_file):
    # Load audio data
    with open(audio_file, "rb") as f:
        audio_data = f.read()

    # Perform speech recognition
    text = deepspeech_model.stt(audio_data)
    return text

# Function to translate text from Hindi to English
def translate_to_english(text):
    # Perform translation
    translated_text = translator.translate(text, src="hi", dest="en").text
    return translated_text

# Directory containing audio files
audio_folder = "path/to/audio/folder"

# Directory to save text files
output_folder = "path/to/output/folder"

# Loop through audio files in the folder
for file in os.listdir(audio_folder):
    if file.endswith(".mp3"):
        audio_file_path = os.path.join(audio_folder, file)

        # Transcribe audio to text
        text = transcribe_audio(audio_file_path)

        # Translate text to English
        english_text = translate_to_english(text)

        # Create output directory if it doesn't exist
        if not os.path.exists(output_folder):
            os.makedirs(output_folder)

        # Save translated text to a text file in the output directory
        output_file_path = os.path.join(output_folder, file.replace(".mp3", "_english.txt"))
        with open(output_file_path, "w", encoding="utf-8") as text_file:
            text_file.write(english_text)

        print(f"Translation saved to: {output_file_path}")
```
