```python
import os
import shutil
import speech_recognition as sr
from langdetect import detect
from google.cloud import translate

def audio_to_text(audio_file):
    recognizer = sr.Recognizer()

    with sr.AudioFile(audio_file) as source:
        audio_data = recognizer.record(source)

    try:
        text = recognizer.recognize_google(audio_data)
        return text
    except sr.UnknownValueError:
        return "Could not understand the audio"
    except sr.RequestError as e:
        return f"Error: {e}"

def translate_text(text, target_language="en"):
    client = translate.TranslationServiceClient()
    parent = client.location_path("YOUR_PROJECT_ID", "global")

    response = client.translate_text(
        parent=parent,
        contents=[text],
        target_language_code=target_language,
    )

    translated_text = response.translations[0].translated_text
    return translated_text

audio_folder = "/content/drive/MyDrive/DARPG-1/DARPG_3/problem_statement_3"  # Replace with your audio folder path
output_folder = "/content/drive/MyDrive/DARPG-1/DARPG_3/Output"  # Output folder for transcribed text

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Iterate over each file in the folder
for filename in os.listdir(audio_folder):
    if filename.endswith(".mp3"):  # Adjust file extension as necessary
        audio_file = os.path.join(audio_folder, filename)

        # Detect language
        try:
            language = detect(audio_to_text(audio_file))
        except:
            language = "unknown"

        # Transcribe and translate based on language
        if language == "en":
            text = audio_to_text(audio_file)
        elif language == "hi":
            text = translate_text(audio_to_text(audio_file), target_language="en")
        else:
            text = "Language detection failed, could not transcribe"

        # Write text to file
        output_file = os.path.join(output_folder, f"{os.path.splitext(filename)[0]}.txt")
        with open(output_file, "w", encoding="utf-8") as f:
            f.write(text)

        print(f"Transcribed and saved text from {filename} to {output_file}")
```
