```python
# Function to transcribe audio segment
def transcribe_audio_segment(audio_segment):
    recognizer = sr.Recognizer()

    with tempfile.NamedTemporaryFile(suffix='.wav', delete=False) as temp_wav:
        audio_segment.export(temp_wav.name, format="wav")

        with sr.AudioFile(temp_wav.name) as source:
            recognizer.adjust_for_ambient_noise(source)
            audio_data = recognizer.record(source)

            try:
                text = recognizer.recognize_google(audio_data)
                return text
            except sr.UnknownValueError:
                print("Google Speech Recognition could not understand audio")
            except sr.RequestError as e:
                print("Could not request results from Google Speech Recognition service; {0}".format(e))
            except ValueError as e:
                print("Error transcribing audio segment")
                print(e)
```

```python
# Path to the folder containing audio files
input_folder = "/content/drive/MyDrive/DARPG/DARPG_3/DARPG_3_WAV_DATASET"
output_folder = "/content/drive/MyDrive/DARPG/DARPG_3/Output_v5"
```

```python
# Ensure output folder exists, create it if not
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
```

```python
# Iterate over each file in the input folder
for file_name in os.listdir(input_folder):
    if file_name.endswith(".mp3"):
        mp3_file = os.path.join(input_folder, file_name)
        wav_file = convert_to_wav(mp3_file, output_folder)
        print("Converted MP3 to WAV:", wav_file)

        audio = AudioSegment.from_wav(wav_file)
        chunk_length_ms = 5000  # 5 seconds
        chunks = [audio[i:i+chunk_length_ms] for i in range(0, len(audio), chunk_length_ms)]

        subtitle_file_path = os.path.join(output_folder, os.path.splitext(file_name)[0] + ".txt")
        with open(subtitle_file_path, "w") as subtitle_file:
            for i, chunk in enumerate(chunks):
                print("Transcribing chunk {}/{} of audio file: {}".format(i+1, len(chunks), file_name))
                transcript = transcribe_audio_segment(chunk)
                if transcript:
                    subtitle_file.write(transcript + "\n")
                else:
                    print("Transcription failed for chunk {}/{} of audio file: {}".format(i+1, len(chunks), file_name))
                    subtitle_file.write("Transcription failed for chunk {}/{}.\n".format(i+1, len(chunks)))

        print("Transcription completed for:", file_name)
```
