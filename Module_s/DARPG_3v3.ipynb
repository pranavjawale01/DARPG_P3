```python
import os
import speech_recognition as sr
from pydub import AudioSegment
from googletrans import Translator

# Function to convert audio file to WAV format
def convert_to_wav(audio_file):
    sound = AudioSegment.from_file(audio_file)
    audio_file_wav = audio_file.replace(".mp3", ".wav")  # Convert to WAV format
    sound.export(audio_file_wav, format="wav")
    return audio_file_wav

# Function to transcribe audio file to text
def transcribe_audio(audio_file):
    recognizer = sr.Recognizer()
    with sr.AudioFile(audio_file) as source:
        audio_data = recognizer.record(source)  # Read the entire audio file
        text = recognizer.recognize_google(audio_data, language="hi-IN")  # Recognize speech using Google Speech Recognition API
    return text

# Function to translate text from Hindi to English
def translate_to_english(text):
    translator = Translator()
    translated_text = translator.translate(text, src="hi", dest="en")
    return translated_text.text

# Directory containing audio files
audio_folder = "/content/drive/MyDrive/DARPG/DARPG_3/problem_statement_3"

# Directory to save text files
output_folder = "/content/drive/MyDrive/DARPG/DARPG_3/Output_v3"

# Loop through audio files in the folder
for file in os.listdir(audio_folder):
    if file.endswith(".mp3"):
        audio_file_path = os.path.join(audio_folder, file)

        # Convert audio file to WAV format
        audio_file_wav = convert_to_wav(audio_file_path)

        # Transcribe audio to Hindi text
        text = transcribe_audio(audio_file_wav)

        # Translate Hindi text to English
        english_text = translate_to_english(text)

        # Create output directory if it doesn't exist
        if not os.path.exists(output_folder):
            os.makedirs(output_folder)

        # Save translated text to a text file in the output directory
        output_file_path = os.path.join(output_folder, file.replace(".mp3", "_english.txt"))
        with open(output_file_path, "w", encoding="utf-8") as text_file:
            text_file.write(english_text)

        print(f"Translation saved to: {output_file_path}")
```
